## Quadratic Approximation

$$
f(x) \approx f(x_0) + f^\prime(x_0)(x-x_0) + \frac{f^{\prime \prime}(x_0)}{2} (x-x_0)^2
$$

### $\ln(1+x)$

$$
\begin{aligned}
f(x) &= \ln(1+x) \\
f^{\prime}(x) &= \frac{1}{1+x} \\
f^{\prime \prime}(x) &= -\frac{1}{(1+x)^2} \\
f(0) &= 0 \\
f^{\prime}(0) &= 1 \\
f^{\prime \prime}(0) &= -1 \\
f(x) &\approx 0 + x - \frac{1}{2} x^2 = x-\frac{x^2}{2} \\
\end{aligned}
$$

### $\cos(x)$

$$
\begin{aligned}
f(x) &= \cos x \\
f^{\prime}(x) &= -\sin x \\
f^{\prime \prime}(x) &= -\cos x \\
f(0) &= 1 \\
f^{\prime}(0) &= 0 \\
f^{\prime \prime}(0) &= -1 \\
f(x) &\approx 1-\frac{x^2}{2} 
\end{aligned}
$$

```
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.linspace(-2,2,100)
plt.plot(x, np.cos(x), label='f')
plt.plot(x, 0*x+1, label='linear approx')
plt.plot(x, 1-x**2/2, label='quadratic approx')
bottom, top = plt.ylim()
plt.vlines(0, bottom, top)
left, right = plt.xlim()
plt.hlines(0, left, right)
plt.legend()
plt.title('cos x and approximations')
plt.show()
```
