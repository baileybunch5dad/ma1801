```
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.linspace(-3,3,100)
def fx(x): return x**2-5
def fpx(x): return 2*x
plt.plot(x,fx(x),label='fx') # graph function
x0=2 # initial guess
def guessplot(xguess,l):
  xrng = np.linspace(xguess-1,xguess+1,100)
  yrng = fpx(xguess)*(xrng-xguess)+fx(xguess)
  plt.plot(xrng,yrng,label=l)
guessplot(x0,'0='+str(x0))  
def fnewt(xn): return xn-fx(xn)/fpx(xn)
x1 = fnewt(x0)
guessplot(x1,'1='+str(x1))
x2 = fnewt(x1)
guessplot(x2,'2='+str(x2))
x3 = fnewt(x2)
guessplot(x3,'3='+str(x3))
x4 = fnewt(x3)
guessplot(x4,'4='+str(x4))
plt.legend()
plt.show()
```

## Newton's Method Formula

$$
x_{n+1} = f(x_n)-\frac{f(x_n)}{f^\prime(x_n)}
$$
