## Prove $\frac{d}{dx} \cos x = -\sin x$

Given
$$
\begin{aligned}
\cos (a+b)&=\cos a \cos b - \sin a \sin b
\end{aligned}
$$

Proof (Derivation)


$$
\begin{aligned}
\frac{d}{dx} \cos (x) &= \lim_{\Delta x \to 0} \frac{\cos (x + \Delta x) - \cos(x)}{\Delta x}  \\
&= \lim_{\Delta x \to 0} \frac{(\cos x \cos \Delta x - \sin x \sin \Delta x) - \cos x}{\Delta x}  \\
&= \lim_{\Delta x \to 0} \left( \cos x \left( \frac{\cos \Delta x - 1}{\Delta x} \right) - \sin x \left( \frac{\sin \Delta x}{\Delta x} \right) \right) \\
&= - \sin x
\end{aligned}
$$


```
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = np.linspace(-7,7,100)
y = np.cos(x)
plt.plot(x,y,color='blue',label=r'$\cos x$')
x0=math.pi/2+.3
y0=math.cos(x0)
plt.scatter(x0,y0, s=10, color='green', marker='o')
plt.text(x0,y0,r'$(x_0,y_0)$')
delta=.5
xs=np.array([x0-delta,x0+delta])
#print(xs)
ys=-np.sin(x0)*(xs-x0)+np.cos(x0)
plt.plot(xs,ys,color='red',label=r'tan=-$\sin x$')

x1=-math.pi/2
y1=math.cos(x1)
plt.scatter(x1,y1, s=10, color='green', marker='o')
plt.text(x1,y1,r'$(x_1,y_1)$')
delta=1
xs=np.array([x1-delta,x1+delta])
#print(xs)
ys=-np.sin(x1)*(xs-x1)+np.cos(x1)
plt.plot(xs,ys,color='green',label=r'tan=-$\sin x$')

plt.legend(loc='best')
plt.show()
```
