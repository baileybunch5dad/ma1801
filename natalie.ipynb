# When Natalie met Sir Isaac

Building a little halfpipe for miniature skateboards what is the slope somewhere up the side. The hafpipe equation is $y=x^2$ and looks like and Natalie wants to find the slope at some arbitrary point $x_0$

```
import matplotlib.pyplot as plt;import numpy as np
def f(x): return x**2
def m(x): return 2*x
x=np.linspace(-1.1,1.1,100);plt.plot(x,f(x))
x0=.5;plt.scatter(x0,f(x0),s=100)
plt.grid();plt.show()
```

We know that the equation for slope is 
$$
\frac{\text{rise}}{\text{run}} = \frac{y_2-y_1}{x_2-x_1} = \frac{f(y_2)-f(y_1)}{x_2-x_1} = m
$$
So if we move a little forward in the $x$ direction from her point $x_0$ by adding a super tiny change, $\Delta$ to $x$


```
x=np.linspace(-1.1,1.1,100);plt.plot(x,f(x))
x0=.5;plt.scatter(x0,f(x0),s=100)
plt.text(x0+.05,f(x0),r'$x_0,f(x_0)$')
x1=x0+.1;plt.scatter(x1,f(x1),s=100)
plt.text(x1+.05,f(x1),r'$x+\Delta,f(x+\Delta)$')
plt.grid();plt.show()
```

Then the slope at the point $x_0$ which we can call $m(x_0)$ is when $\Delta$ gets smaller and smaller, eventually hitting infinitely close to 0, so for all practical purposes 0, so the slope $m$ at the point $x_0$ is
$$
m \text{ at point }x_0=
\frac{\text{rise}}{\text{run}} = 
\frac{f(x_0+\Delta)-f(x_0)}{x_0+\Delta - x_0} =
\frac{(x_0+\Delta)^2-x_0^2}{x_0+\Delta - x_0} =
\frac{x_0^2+2x\Delta+\Delta^2-x_0^2}{\Delta} =
\frac{2x\Delta+\Delta^2}{\Delta} \\
$$
But we have $\Delta$ in the denominator which is $\approx0$ so multiply both sides by $\frac{1}{\Delta}$
$$
m(x_0) =
\frac{\frac{1}{\Delta}}{\frac{1}{\Delta}} \frac{2x_0\Delta+\Delta^2}{\Delta} =
\frac{2x_0+\Delta}{1}=
2x_0+\Delta=
\boxed{2x_0}
$$
Now we can use this to draw the tangent line
$$
\frac{y-y_0}{x-x_0} = m(x_0) 
$$

```
import matplotlib.pyplot as plt;import numpy as np
x=np.linspace(-1.1,1.1,100)
plt.plot(x,f(x))
x0=.5;plt.scatter(x0,f(x0),s=100)
x=np.linspace(.1,1,10)
y=m(x0)*(x-x0)+f(x0)
plt.plot(x,y)
plt.grid();plt.show()
```

## What is the x-intercept of the tangent line at her chosen point $x_0$

The equation of the x-intercept is $y=0$ so

$$
\begin{aligned}
y - y_0 &= m(x_0) (x-x_0)  \\
y-f(x_0) &= m( x_0) (x-x_0) \\
-f(x_0) &= m(x_0) (x-x_0) & y=0 \\
-\frac{f(x_0)}{m(x_0)} &= x - x_0 \\
x &= \boxed { x_0 - \frac{f(x_0)}{m(x_0)} } \\
\end{aligned}
$$

## Newton's Method: Keep on drawing tangents and finding the x intercept to solve square roots

$$
\begin{aligned}
x^2 &= 5 \\\
\implies y&=x^2 - 5 \implies f(x)=x^2-5 \\
m(x_0) &= \frac{f(x_0+\Delta)-f(x_0)}{x_0+\Delta-x_0} = \frac{(x_0+\Delta)^2-5-x_0^2-5}{\Delta} = \boxed{ 2x_0 } \\
\end{aligned}
$$

So start with an initial guess $x_0$ every subsequent guess $n=1,2,3,\ldots$  is the x intercept of the tangent of the previous guess

$$
\begin{aligned}
x_{n+1} &= x_n - \frac{f(x_n)}{m(x_n)} \\
x_{n+1} &= x_n - \frac{x_n^2-5}{2x_n} \\
x_{n+1} &= x_n - \frac{1}{2} x_n + \frac{5}{2 x_n} = \boxed{ \frac{x_n}{2} + \frac{5}{2x_n} } \\
\end{aligned}
$$

Plugging in the numbers
$$
\begin{aligned}
x_0 &= \boxed{ 2 } &\text{initial guess} \\
x_1 &= \frac{x_n}{2} + \frac{5}{2x_n} \\
&= \frac{2}{2}+\frac{5}{2(2)} \\
&= \boxed{ \frac{9}{4} } = 2.25 & \sqrt{5}\text{ to 1 decimal place} \\
x_2 &= \frac{\frac{9}{4}}{2} + \frac{5}{2\frac{9}{4}} \\
&= \frac{9}{8} + \frac{20}{18} \\
&= \frac{9\cdot 18 + 20\cdot 8}{8 \cdot 18} \\
&= \frac{162+160}{144} = \boxed{ \frac{322}{144} } = 2.236 & \sqrt{5}\text{ to 3 decimal places!!} \\
x_3 &= \frac{\frac{322}{144}}{2}  + \frac{5}{2\frac{322}{144}} \\
&= \frac{322}{2\cdot 144} + \frac{5\cdot 144}{2 \cdot 322} \\
&= \frac{322\cdot 322 + 5\cdot 144 \cdot 144}{2 \cdot 144 \cdot 322} \\
&= \boxed{ \frac{207364}{92736}  } = 2.236067977 & \sqrt{5}\text{ to 9 decimal places!!} \\ 
\end{aligned}
$$
