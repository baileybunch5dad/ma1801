# Find area beteen $x=y^2$ and $y=x-2$

First graph it

```
import matplotlib.pyplot as plt
import numpy as np
a=0
b=3
def f(y): return y*y
y = np.linspace(-2,2)
x = f(y)
plt.plot(x,y)
def g(x): return x-2
x =np.linspace(0,4)
y = g(x)
plt.plot(x,y)
left,right=plt.xlim()
bot,top=plt.ylim()
plt.hlines(0,left,right,color='black')
plt.vlines(0,bot,top,color='black')
plt.show()
```

FInd the intersection points

(0,0) is where the parabola intersects the x-axis

Now where does $x=y^2$ intersect $y=x-2$

$$
\begin{aligned}
y&=y^2-2 \\
y^2-y-2&=0 \\
(y-2)(y+1)&= 0 \\
y &= 2,y = -1 \\
y &= x-2 \\
x &= y+2 \\
\text{points} &= [4,2], [1,-1] \\
\end{aligned}
$$

So the integral is in two parts, top is always $\sqrt{x}$, the left part bounded by the bottom part of the curve ($-\sqrt{x}$), and the right part $x-2$

$$
\begin{aligned}
\text{Area} &= \int_0^1 [(\sqrt{x}) - (-\sqrt{x})]dx + \int_1^4 [(\sqrt{x})-(x-2)]dx \\
&= \int_0^1 2\sqrt{x} dx + \int_1^4 \sqrt{x} - x + 2 dx \\
&=  \left( \frac{4}{3}x^{3/2} \right) \bigg\vert_0^1 +  \left( \frac{2}{3} x^{3/2}-x^2/2 + 2x \right) \bigg\vert_1^4   \\
&= (4/3-0) + ((2/3\cdot 8 - 8 + 8) - (2/3 - 1/2 + 2)) \\
&= 4/3 + 16/3 - ( 2/3 - 1/2 + 2 ) \\
&= 20/3  - (4/6-3/6+12/6) \\ 
&= 40/6 - 13/6 \\
&= 9/2 \\
&= \boxed{ 4.5 } \\
\end{aligned}
$$
