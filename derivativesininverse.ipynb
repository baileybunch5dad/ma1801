## Derivative of $\sin^{-1} x$

$$
\begin{aligned}
y &= \sin^{-1} x \\
\sin y &= x \\
(\cos y )(y^\prime) &= 1 & \text{chain rule w implicit diff}\\
y^\prime &= \frac{1}{\cos y} \\
&= \frac{1}{\cos (\sin^{-1} x)} \\
&= \frac{1}{\sqrt{1-x^2}} \\
\end{aligned}
$$

```
import matplotlib.pyplot as plt
import numpy as np
import math

def ff(x): return np.arcsin(x)
def fp(x): return 1/np.cos(np.arcsin(x)) 

x = np.linspace(-.99,.99,200)
plt.plot(x,ff(x),label=r'$\sin^{-1} x$')

def dot(x0,y0,hor='center',ver='center'):
  plt.scatter(x0,y0, s=10, color='green', marker='o')
  txt = "({:.2f},{:.2f})".format(x0,y0)
  plt.text(x0,y0,txt,ha=hor,va=ver)

x0=-.8
y0=ff(x0)
dot(x0,y0,'right','bottom')
delta=.2
xs = np.array([x0-delta,x0+delta]) 
# instantaneous slope at x0
m = fp(x0)
ys = m * (xs-x0) + y0
plt.plot(xs,ys,label=r'$\frac{d}{dx} =1/\sqrt{1-x^2}$')

x0=.9
y0=ff(x0)
dot(x0,y0,'right','bottom')
delta=.15
xs = np.array([x0-delta,x0+delta]) 
# instantaneous slope at x0
m = fp(x0)
ys = m * (xs-x0) + y0
plt.plot(xs,ys,label=r'$\frac{d}{dx} =1/\sqrt{1-x^2}$')


plt.legend()
plt.show()
```
